```python
import numpy as np
import scipy as sp
from tkinter import filedialog
import matplotlib.pyplot as plt
from scipy import signal
import os
```

```python
fnames = filedialog.askopenfilenames()
```

```python
import pickle as pkl
import copy

spec_list = []
g_spec_list = []
for fname in fnames:
    with open(fname, 'rb') as f:
        d = pkl.load(f)
        spec_list.append(d['absorption']['absorbance'])
        g_spec_list.append(d['PL']['gain_spectrum'][1])

org_list = copy.deepcopy(spec_list)
        
len(spec_list)
    
```

```python
calc_range = [300,800]
c_range = [np.where(spec_list[0][0] > calc_range[0])[0][0], np.where(spec_list[0][0] < calc_range[1])[0][-1]]
c_range
```

```python
filter_size = 50
fil = np.ones(filter_size)/filter_size

for spec in spec_list: 
    spec = spec[:, c_range[0]:c_range[1]+1]
    spec[1] = signal.convolve(spec[1], fil, mode='same')
for spec in g_spec_list:
    spec = spec[c_range[0]:c_range[1]+1]
```

```python
threshold = -0.0005
b = 200
c = 25

end_indices = []
end_wavelengths = [] 

for j, spec in enumerate(spec_list):
    end_indices.append(None)
    end_wavelengths.append(None) 
    
    grad = np.gradient(spec[1])/np.max(spec[1])/(spec[0][1] - spec[0][0])
    
    indices = np.where(grad < threshold)[0]
    
    for i in range(len(indices)-c):
        b = min(b, indices[-i-1])
        if indices[-i-1] - indices[-i-1-c] == c and (spec[1][indices[-i-1]-b:indices[-i-1]] > 0).all():
            end_indices[j] = indices[-i-1]
            end_wavelengths[j] = spec[0][end_indices[j]]
            break

    print(os.path.basename(fnames[j]), end_indices[j], end_wavelengths[j], g_spec_list[j][end_indices[j]])
```

```python
for i, spec in enumerate(org_list):
    plt.plot(spec[0], spec[1])
    plt.plot([200,1000], [0,0])
    plt.title(os.path.basename(fnames[i]))
    plt.vlines(end_wavelengths[i], -0.2 * np.max(spec[1]), 1.2 * np.max(spec[1]), color = 'r')
    plt.show()
```
